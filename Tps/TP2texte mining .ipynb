```python

words_set = set()
 
for doc in  df:
    words = doc.split(' ')
    words_set = words_set.union(set(words))
     
print('Number of words in the df:',len(words_set))
print('The words in the df: \n', words_set)
```

Computing Term Frequency

```python

n_docs = len(df)         #Â·Number of documents in the df
n_words_set = len(words_set) #Â·Number of unique words in the df
 
df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=words_set)
 
# Compute Term Frequency (TF)
for i in range(n_docs):
    words = df[i].split(' ') # Words in the document
    for w in words:
        df_tf[w][i] = df_tf[w][i] + (1 / len(words))
         
df_tf
```

Computing Inverse Document Frequency
inverse document frequency (IDF)

```python


print("IDF of: ")
 
idf = {}
 
for w in words_set:
    k = 0    # number of documents in the df that contain this word
     
    for i in range(n_docs):
        if w in df[i].split():
            k += 1
             
    idf[w] =  np.log10(n_docs / k)
     
    print(f'{w:>15}: {idf[w]:>10}' )
```

Putting it Together: Computing TF-IDF

```python

df_tf_idf = df_tf.copy()
 
for w in words_set:
    for i in range(n_docs):
        df_tf_idf[w][i] = df_tf[w][i] * idf[w]
         
df_tf_idf
```

```python
X=data['Text']
Y=data['IsToxic']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=123)
```

```python
print('Training Data :', X_train.shape)

print('Testing Data : ', X_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X_train_cv = cv.fit_transform(X_train)

X_train_cv.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train_cv, y_train)




# transform X_test using CV

X_test_cv = cv.transform(X_test)




# generate predictions

predictions = lr.predict(X_test_cv)

predictions
```

```python
from sklearn import metrics

classification = pd.DataFrame(metrics.confusion_matrix(y_test,predictions), index=['True','False'], columns=['True','False'])

classification 
```


# Data

```python
data = pd.read_csv('youtoxic_english_1000.csv')
data
```

# Clean data

```python
data['Text'] = data['Text'].str.lower()
data
```

```python
data['Text'] = data['Text'].str.strip()
data
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
lemmatize=nltk.WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('omw-1.4')
def CleanData(Text):

    no_punct  = ""
    for char in Text:
        if char not in string.punctuation:
            no_punct = no_punct + char
    no_punct  = ''.join(no_punct )
    no_punct  = [lemmatize.lemmatize(word) for word in no_punct .split()]
    no_punct  = ' '.join(no_punct ) 
    return [word.lower() for word in no_punct.split() if word.lower() not in stopwords.words('english')]
```

```python
data['Text'].head(5).apply(CleanData)
```

## Training a model

```python
from sklearn.model_selection import train_test_split

text_train, text_test, toxic_train, toxic_test = \
train_test_split(df['Text'], df['IsToxic'], test_size=0.2)

```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=CleanData)),  
    ('tfidf', TfidfTransformer()),  
    ('Classification', MultinomialNB()),  
])
```

```python
pipeline.fit(text_train,toxic_train)
```

```python
predictions = pipeline.predict(text_test)
```

```python
print(classification_report(predictions,toxic_test))
```
